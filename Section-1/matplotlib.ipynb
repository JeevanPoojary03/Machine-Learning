```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

# create a line plot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Basic line plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
# customized line plot

plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=2,markersize=9)
plt.grid(True)
```

```python
# Multiple Plots
# sample data

x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(x,y2,color='purple')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(y1,y2,color='red')
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(y1,y1,color='blue')
plt.title("Plot 4")
```

```python
# bar plot
categories=['A','B','C','D','E']
values=[5,7,3,8,6]

# create a bar plot
plt.bar(categories,values,color='purple')
```

```python
data=[1,2,2,2,3,3,3,4,4,4,5,5,5,5,5,]

plt.hist(data,bins=5,color='orange')
```

```python
# create a scatter plot
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.scatter(x,y,color='blue',marker='x')
```

```python
# pie chart

labels=['A','B','C','D']
sizes=[30,20,40,10]
color=['gold','yellowgreen','lightcoral','blue']
explode=(0.2,0,0,0) #move out the first slice

plt.pie(sizes,labels=labels,colors=color,shadow=True)
```
